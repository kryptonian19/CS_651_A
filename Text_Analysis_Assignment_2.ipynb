# Tokenisation using regex

```python
import regex as re

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

testtext = "I'm going to become the king of Pirates! https.strawhats.com"
testtokens = tokenize(testtext)
 
print(testtokens)
```

# Stop words

```python
import nltk 
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

# Removing Stop Words

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

# Making a pipeline

```python
#This step creates a pipeline
pipeline = [str.lower, tokenize, remove_stop]

def prepare(case_text, pipeline):
    tokens = case_text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

# Running the pipeline

```python
#This step runs the pipeline
df['tokens'] = df['case_text'].apply(prepare, pipeline=pipeline)
```

# Counting Tokens

```python
#This step counts the number of tokens
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

# Counter()

```python
from collections import Counter

tokens = tokenize("The Straw hat belongs to luffy!")

testcount= Counter(tokens)
print(testcount)
```

# Adding new data to current count

```python
more_tokens = tokenize("The Straw hat is given to luffy by shanks!")
testcount.update(more_tokens)
print(testcount)
```

# Updating count on dataset

```python
#This step updates the counter
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

# Word Frequency

```python
#This step return words with freq greater than 1
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

# Sorting and displaying word frequencies

```python
#This step sorts and displays the word frequencies
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

# Describing word frequencies

```python
#This step describes the word frequencies
freq_df.describe()
```

# Wordcloud

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt

case_text = df.query("case_outcome=='cited'")['case_text'].values[0]

plt.figure(figsize=(14, 7)) ###

#This step is used to generate word cloud
wc = WordCloud(max_words=100, stopwords=stopwords)
wc.generate(case_text)

plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```

# Wordcloud Continuation

```python
wc = WordCloud(background_color='white', width = 300, height=300, max_words=100, margin=2)
wc.fit_words(freq_df['freq'].to_dict() )
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
