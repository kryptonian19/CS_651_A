# Tokenisation using regex

```python
import regex as re

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)

testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
 
print(testtokens)
```

# Stopwords

```python
import nltk 
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

# Removing Stop Words

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

# Making a pipeline

```python
pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

# Running the pipeline

```python
df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

# Counting Tokens

```python
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

# Counter()

```python
from collections import Counter

tokens = tokenize("She likes my cats and my cats like my sofa.")

testcount= Counter(tokens)
print(testcount)
```

# Adding new data to current count

```python
more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)
```

# Updating count

```python
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

# Word Frequency

```python
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

# Sorting and Displaying word frequencies

```python
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

# Describing word frequencies

```python
freq_df.describe()
```

# Wordcloud

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt

text = df.query("year==2015 and country=='USA'")['text'].values[0]

plt.figure(figsize=(15, 15)) ###
wc = WordCloud(width = 960, height=540, max_words=100, stopwords=stopwords)
wc.generate(text)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```

# Word cloud continuation

```python
wc = WordCloud(background_color='white', width = 960, height=540, max_words=100, margin=2)
wc.fit_words(freq_df['freq'].to_dict() )
plt.figure(figsize = (15, 15), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
